```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("customer_support_knowledge_base.pdf")

documents = loader.load()

print("Total Pages:", len(documents))
```

## chunk split

```python
!pip install langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Create text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

# Split documents
chunks = text_splitter.split_documents(documents)

print("Total Chunks Created:", len(chunks))
```

## create  Embeddings + Store in ChromaDB

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings

# Create embedding model
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Store chunks into ChromaDB
vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=embedding_model
)

print("Vector Database Created Successfully")
```

## Retriever

```python
retriever = vectorstore.as_retriever()

print("Retriever Ready")
```

## LLM + Query Answer System

```python
def get_context(query):
    docs = retriever.invoke(query)

    # take only top 2 chunks
    return "\n".join([d.page_content for d in docs[:2]])
```

```python
def generate_answer(query):

    context = get_context(query)

    # break into lines
    lines = context.split("\n")

    # keep only meaningful lines
    clean_lines = []

    for l in lines:
        if len(l) > 25 and "Customers" in l or "Order" in l or "Refund" in l or "Cancel" in l:
            clean_lines.append(l.strip())

    # fallback if filter is empty
    if not clean_lines:
        clean_lines = [l for l in lines if len(l) > 20]

    return "\n".join(clean_lines[:5])
```

```python
print(generate_answer("How do I cancel my order?"))
```

```python
print(generate_answer("what is Refund Timeline?"))
```

```python
print(generate_answer("What is refund policy?"))
```
